## Problem Statement

In this program we show an example of using dictionaries to keep track of information in a phonebook.

## Solution

```python
from typing import Dict

phonebook: Dict = {}
def add_contact() -> Dict[str, str]:                   

    while True:
        name: str = input("Name: ")

        number: str = input("Number: ")
        phonebook[name]: str = number
        print(f"Added {name} with number {number} to the phonebook.")
        
        qs = input("Press Enter to add another contact or type 'e' to exit.")
        if qs == "e":
            break
        
    main()
    return phonebook


def view_contacts():

    for name in phonebook:
        print(f"{name} -> {phonebook[name]}")
    main()


def search_contact():

    while True:
        name: str = input("Enter name to lookup: ")
   
        if name == "":
            print("Please enter a valid name.")
            continue
        
        if name not in phonebook:
            print(name + " is not in the phonebook")
        else:
            print(f"{name} -> {phonebook[name]}")
            
        qs = input("Press Enter to search for another contact or type 'e' to exit.")
        if name == "e":
            break
        
        main()

def main():
    print("\n\nWelcome to the phonebook application!")
    print("1. Add a new contact")
    print("2. View all contacts")
    print("3. Search for a contact")
    print("4. Exit")
    print("Please select an option (1-4):")

    user_choice = input()
    
    if user_choice == "1":
        phonebook = add_contact()
    elif user_choice == "2":
        view_contacts()
    elif user_choice == "3":
        search_contact()
    elif user_choice == "4":
        print("Exiting the application.")
        exit()
    else:
        print("Invalid choice. Please try again.")
        main()
        
        
if __name__ == '__main__':
    main()
```
