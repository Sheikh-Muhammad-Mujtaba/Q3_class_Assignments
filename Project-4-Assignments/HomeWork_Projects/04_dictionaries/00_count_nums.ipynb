## Problem Statement

This program counts the number of times each number appears in a list. It uses a dictionary to keep track of the information.

An example run of the program looks like this (user input is in blue):

Enter a number: 3
Enter a number: 4
Enter a number: 3
Enter a number: 6
Enter a number: 4
Enter a number: 3
Enter a number: 12
Enter a number: 
3 appears 3 times.
4 appears 2 times.
6 appears 1 times.
12 appears 1 times.

## Solution


```python
from typing import Dict, List

def get_user_numbers() -> List[int]:

    user_numbers: List[int] = []
    
    while True:
        user_input: int = input("Enter a number: ")
        
        if user_input == "":
            break
        
        try:
            num: int = int(user_input)
            user_numbers.append(num)
            
        except ValueError:
            print("Please enter a valid number.")
            
            
    return user_numbers

def count_nums(num_lst) -> Dict[int, int]:

    num_dict: Dict = {}
    
    for num in num_lst:
        if num not in num_dict:
            num_dict[num]: int = 1
        else:
            num_dict[num] += 1
    
    if len(num_dict) == 0:
        print("No numbers were entered.")

    return num_dict


def print_counts(num_dict):

    for num in num_dict:
        print(f"{num} appears {num_dict[num]} times.")


def main():
    user_numbers: int = get_user_numbers()
    num_dict: Dict = count_nums(user_numbers)
    print_counts(num_dict)

if __name__ == '__main__':
    main()
```
